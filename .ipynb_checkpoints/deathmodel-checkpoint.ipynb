```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,GridSearchCV
```

```python
df = pd.read_excel("./datasets/COVID-19-geographic-disbtribution-worldwide-2020-03-22.xlsx")
```

```python
df.head()
```

```python
df = df.iloc[:,:8]
```

```python
df
```

```python
df.DateRep = pd.to_datetime(df.DateRep)
```

```python
df
```

```python
df.sort_values(["DateRep","Countries and territories"], inplace = True, ascending = True)
```

```python
df
```

```python
df["cumdeaths"] = df.groupby("Countries and territories").Deaths.cumsum()
df
```

```python
df["cumcases"] = df.groupby("Countries and territories").Cases.cumsum()
df
```

```python
pd.set_option("display.max_rows", 500)
```

```python
df[df["Countries and territories"] == "United_Kingdom"]
```

```python
def dffilter(df):
    df = df[df.cumdeaths > 0]
    return df

df = df.groupby("Countries and territories", as_index = False).apply(lambda x : dffilter(x))
```

```python
df = pd.get_dummies(df,columns = ["Countries and territories"])
```

```python
df.reset_index(inplace = True)
```

```python
df.drop(columns = ["level_0", "level_1"], inplace = True)
```

```python
df
```

```python
def count_days(df):
    num_rows = len(df)
    rows = [i for i in range(1, num_rows+1, 1)]
    df["daycount"] = rows
    return df
```

```python
df=df.groupby("GeoId").apply(lambda x:count_days(x))
```

```python
df.columns
```

```python
delcols=['DateRep', 'Day', 'Month', 'Year', 'Cases', 'Deaths', 'GeoId', 'cumcases']
```

```python
dfmodel = df.drop(columns=delcols)
```

```python
list(dfmodel)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
X = dfmodel.iloc[:,1:]
y = dfmodel.iloc[:,0]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
rf=RandomForestRegressor()
```

```python
params = {"n_estimators": [50,100,150,200],
         "max_depth": [5,10,15,20], 
         "min_samples_split": [2,3,5]}
```

```python
cv = GridSearchCV(rf, params, cv=10, n_jobs=-1, scoring="r2", verbose=1)
```

```python
model=cv.fit(X_train, y_train)
```

```python
X_test["predictions"]=model.predict(X_test)
```

```python
X_test["cumdeaths"]= y_test
```

```python
plt.figure(figsize=(12,12))
plt.scatter(X_test.predictions, X_test.cumdeaths)
plt.plot(X_test.cumdeaths,X_test.cumdeaths,color="black", linestyle="--")
plt.xlabel("predicted deaths")
plt.ylabel("actual deaths")
plt.show()
```
