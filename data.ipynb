```python
import pandas as pd
import requests
```

```python
url = "https://www.worldometers.info/coronavirus/#countries"

header = {"User-Agent": "COVID_19"}
```

```python
r = requests.get(url, headers=header)
```

```python
df = pd.read_html(r.text)
```

```python
df = pd.concat(df, axis=0)
```
